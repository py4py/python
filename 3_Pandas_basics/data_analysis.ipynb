### ADVANCED FUNCTIONS

```python
# create new DataFrame with one column changing datatype 'object' to 'string' modifying string and seting type to 'int'
# pandas.DataFrame - creates new DataFrame
clean_data = pandas.DataFrame(data['Nowe przypadki'].astype(str).str.replace("+","").astype(int))
clean_data
```

```python
# add new column to new DataFrame makeing column type of 'datetime'
# pandas.to_datetime - converts to datetime
clean_data['date'] = pandas.to_datetime((data['Data'].astype(str) + '.20')
                                        .str.replace(".","-")
                                        .str.replace("-1-","-10-")
                                        .apply(lambda i: '{0:0>8}'.format(i)), format='%d-%m-%y')
clean_data
```

```python
# add new column with rolling and basic statistic function on top
# .rolling - can be applied on a series of data; basicly takse last 100 records (window parametr) of column and applies mean() (min_periods - min number of records that needs value)
clean_data['rolling_mean'] = clean_data['Nowe przypadki'].rolling(window=100, min_periods=0).mean()
clean_data
```

```python
# add column with exponential weights (might be ussed for smoothing the curve of records)
clean_data['ewm'] = clean_data['Nowe przypadki'].ewm(com=0.50).mean()
clean_data
```
